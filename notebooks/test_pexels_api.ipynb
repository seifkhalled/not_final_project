# Test Pexels API Key

This notebook tests if the Pexels API key from the .env file is working correctly.

```python
# Import required libraries
import os
from dotenv import load_dotenv
import requests

# Load environment variables from .env file
load_dotenv()

# Get Pexels API key
pexels_api_key = os.getenv('PEXELS_API_KEY')
print(f"Pexels API Key found: {pexels_api_key is not None}")
if pexels_api_key:
    print(f"API Key (first 10 chars): {pexels_api_key[:10]}...")
else:
    print("ERROR: PEXELS_API_KEY not found in environment variables")
```

```python
# Test the API key by making a request to Pexels API
if pexels_api_key:
    url = "https://api.pexels.com/v1/search"
    headers = {
        "Authorization": pexels_api_key
    }
    params = {
        "query": "pyramids",
        "per_page": 1
    }

    try:
        response = requests.get(url, headers=headers, params=params)
        print(f"Status Code: {response.status_code}")
        
        if response.status_code == 200:
            data = response.json()
            print("SUCCESS: Pexels API is working!")
            print(f"Total results found: {data.get('total_results', 0)}")
            if data.get('photos'):
                print(f"First photo: {data['photos'][0]['src']['original']}")
        else:
            print(f"ERROR: API request failed with status {response.status_code}")
            print(f"Response: {response.text}")
    except Exception as e:
        print(f"ERROR: Exception occurred: {str(e)}")
else:
    print("Skipping API test - no API key found")
```
